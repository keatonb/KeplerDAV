# Kepler DAV SED

Here I attempt to construct a spectral energy distribution from available magnitudes of the Kepler DAV KIC 4552982.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Collect all available photometry and plot

```python
#It is not clear that these values are new, but I had not (to my memory) previously seen KIC magitudes available
#for KIC 4552982.  According to the discussion at http://archive.stsci.edu/kepler/kepler_fov/explanations.html
#KIC may be more trustworthy than KIS, and they are already in the AB system.
g_kic=18.050
r_kic=17.797
gi_kic=0.283
i_kic=g_kic-gi_kic

print g_kic-r_kic
print r_kic-i_kic

Kp_mag= 0.25*g_kic + 0.75*r_kic #Kepler magnitude calculation equation?

#Rough Uncertainties from Brown et al. 2011, AJ, 142, 112
g_kic_err=0.07
r_kic_err=0.06
i_kic_err=0.06


#Filter widths:
#g r i from Fukugita et al., 1996, AJ 111, 1748
g_kic_HWHM=700. #angstroms
r_kic_HWHM=700.
i_kic_HWHM=750.
```

```python
#Plot model spectrum near spectroscopic parameters to check reliablility of magnitudes
spec=np.loadtxt('dai10750_825.dk',skiprows=38)
spec[:,1]=-2.5*np.log10(spec[:,1]*(spec[:,0]**2.))#to magnitude scale
```

```python
#Plot points with errors
#Horizontal error bars are bandwidths for reference.
plt.errorbar([4846.,6240,7743],[g_kic,r_kic,i_kic],yerr=[g_kic_err,r_kic_err,i_kic_err],
             xerr=[g_kic_HWHM,r_kic_HWHM,i_kic_HWHM],fmt='o')

plt.plot(spec[:,0],spec[:,1]+74.6) #plus approximate constant to put on same scale.
plt.ylim(19,17)
plt.xlim(2000,10000)
plt.xlabel('wavelength (ang)')
plt.ylabel('magnitude (AB)')
```

```python
#There is also a J-band magnitude from UKIRT: http://keplerscience.arc.nasa.gov/ToolsUKIRT.shtml
#(I verified correct source in image)
#Using jAperMag3
J_IRT=17.764 #Vega magnitude!
J_IRT=J_IRT+0.938 #to AB mag; Hewett 2006, MNRAS, 367, 454
J_IRT_err=0.03
J_IRT_wave=12483.#Hewett 2006, MNRAS, 367, 454
J_IRT_width=1065.
```

```python
#Plot points with errors and filter widths.
plt.errorbar([4846.,6240,7743,J_IRT_wave],[g_kic,r_kic,i_kic,J_IRT],yerr=[g_kic_err,r_kic_err,i_kic_err,J_IRT_err],
             xerr=[g_kic_HWHM,r_kic_HWHM,i_kic_HWHM,J_IRT_width],fmt='o')

plt.plot(spec[:,0],spec[:,1]+74.6)
plt.ylim(19,17)
plt.xlim(2000,14000)
plt.xlabel('wavelength (angstroms)')
plt.ylabel('magnitude (AB)')
```

```python
#Kepler-INT survey magnitudes (Greiss et al.)
intmags=np.array([17.362,17.755,17.677,17.565])#Ugri, Vega system
interrs=np.array([0.007,0.005,0.007,0.009])
intwave=np.array([3581.,4846.,6240,7743])#filter center in angstroms

#These can be converted to a AB magnitude system using the corrections from  
#http://www.astronomy.ohio-state.edu/~martini/usefuldata.html
adjustint=np.array([0.79,-0.08,0.16,0.37]) #Ugri (is this reliable?)

#correct these magnitudes
intabmags=intmags+adjustint
```

```python
#Synthetic absolute magnitudes from Bergeron's cooling model: DA teff=11000,M=0.7
bergeron=np.array([12.101,12.221,12.396,12.172+0.938]) #griJ
bergeronwave=[4670.,6156.,7471.,12350.]

#Vega magnitudes from Holberg & Bergeron 2006
#griJ
#(is this right?)
vega=np.array([-0.1025,0.1460,0.3656,0.023])

#Apply these corrections
bergeron=bergeron-vega
```

```python
#Plot all magnitudes

#Plot KIC, UKIRT data points with errors
plt.errorbar([4846.,6240,7743,J_IRT_wave],[g_kic,r_kic,i_kic,J_IRT],yerr=[g_kic_err,r_kic_err,i_kic_err,J_IRT_err],fmt='o')
#Plot corrected KIS magnitudes with errors
plt.errorbar(intwave,intabmags,yerr=interrs,fmt='o',color='g')
#Plot bergeron calculated
plt.scatter(bergeronwave,bergeron-bergeron[2]+i_kic,color='red') #anchor at i_kic

plt.plot(spec[:,0],spec[:,1]+74.8)
plt.ylim(19,17)
plt.xlim(2000,13000)
plt.xlabel('wavelength (angstroms)')
plt.ylabel('magnitude (AB)')
```
